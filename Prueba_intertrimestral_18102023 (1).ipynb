## Ejercicio 1 (2 puntos):

a) Crea una función que calcule y devuelva el factorial de un número entero. **(0.6 puntos)**

b) Crea una función que verifique si un número es primo o no. **(0.6 puntos)**

c) Muestra en un dataframe los 50 primeros números positivos, si es primo y su factorial utilizando las funciones anteriores. **(0.6 puntos)**

d) ¿Cómo se podría programar en una clase las tres operaciones anteriores? **(0.2 puntos)**

```python
# funcion factorial de un número: producto de los naturales anteriores a él  por él
def factorial(numero):
    if numero < 0:
        return "No existe factorial para negativos"
    elif numero == 0:
        return 1
    else:
        resultado = 1
        for i in range(1, numero + 1):
            resultado *= i
        return resultado
```

```python
# Función número primo o no : Los números primos son aquellos que solo son divisibles entre ellos mismos y el 1
# Si intentamos dividirlos por cualquier otro número, el resultado no es entero. Usamos operador módulo: %

def es_primo(num):
    for n in range(2, num):
        if num % n == 0:
            return False
    return True
```

```python
# Dataframe que tenga los primeros 50 números, indique si es primo y de su factorial.

data = []
for i in range(1, 51):
    is_prime = es_primo(i)
    fact = factorial(i)
    data.append([i, is_prime, fact])

# Crear un DataFrame a partir de la lista de datos
df = pd.DataFrame(data, columns=['Número', 'Es Primo', 'Factorial'])

# DataFrame
df.head()
```

```python
#¿Cómo se podría programar en una clase las tres operaciones anteriores?

class Ejercicio1:
    def __init__(self):
        self.data = []

    def factorial(self, numero):
        if numero < 0:
            return "No existe factorial para negativos"
        elif numero == 0:
            return 1
        else:
            resultado = 1
            for i in range(1, numero + 1):
                resultado *= i
            return resultado

    def es_primo(self, num):
        for n in range(2, num):
            if num % n == 0:
                return False
        return True

    def calcular_y_crear_data_frame(self, start, end):
        columns=['Número', 'Es Primo', 'Factorial']
        for i in range(start, end + 1):
            is_prime = self.es_primo(i)
            fact = self.factorial(i)
            self.data.append([i, is_prime, fact])

        df = pd.DataFrame(self.data, columns=columns)
        return df
```

```python
#A) extraer y transformar.
from sklearn.datasets import fetch_california_housing

california_housing = fetch_california_housing()
df = pd.DataFrame(data = california_housing['data'], columns = california_housing["feature_names"])
df["target"] = california_housing["target"]
df.head()
```

```python
#B ) Función que muestra cosas del dataset.
def info_dataset(dataframe):
  #suma de NAs
  na_suma = df.isna().sum()
  col_types = df.dtypes
  col_descripciones = df.describe()
  info_df = pd.DataFrame({'NAs': na_suma, 'Data Type': col_types}).join(col_descripciones.transpose())

  return info_df
```

```python
import pandas as pd
import numpy as np
import sklearn.datasets
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.metrics import (classification_report, mean_squared_error, mean_absolute_error, mean_absolute_percentage_error,
                            silhouette_score, r2_score)
from sklearn.linear_model import ElasticNet
from sklearn.cluster import KMeans
import plotly.express as px
```

```python

from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
```

```python
# C ) Regresión lineal y random forest que predigan median house value
X = california_housing["data"]
y = california_housing["target"]
X_train, X_test, y_train, y_test = train_test_split(X, y ,test_size = 0.3, random_state = 123)



#REGRESION
reg = LinearRegression()
reg.fit(X_train, y_train)
y_pred = reg.predict(X_test)
mse_reg = mean_squared_error(y_test, y_pred)
mae_reg = mean_absolute_error(y_test, y_pred)
r2_reg = r2_score(y_test, y_pred)


# Random Forest
rf_model = RandomForestRegressor(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)
y_pred = rf_model.predict(X_test)
mse_rf = mean_squared_error(y_test, y_pred)
mae_rf = mean_absolute_error(y_test, y_pred)
r2_rf = r2_score(y_test, y_pred)
```

```python
#COEFICIENTES REGRESIÓN
importances = reg.coef_
column_names = df.columns[:-1].tolist()

feature_importances = {}
for i in range(len(column_names)):
    feature_name = column_names[i]
    importance = importances[i]
    feature_importances[feature_name] = importance

data = pd.DataFrame({'Feature': list(feature_importances.keys()), 'Coefficient': list(feature_importances.values())})

# Crear el histograma utilizando Plotly Express
fig = px.bar(data, x='Feature', y='Coefficient', title='#COEFICIENTES REGRESIÓN')

fig.update_xaxes(title_text='Características')
fig.update_yaxes(title_text='Coeficiente')

fig.show()


```
